#### I used this three twitter accounts to show how the tweets are considered by the people like positive, negative or neutral, neutral is to say that people who saw those tweets may not react about that particular tweet.


```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')
sentiments = [get_sentiments(tweet_modi.text) for tweet_modi in modi_tweets]
# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])
ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))
ax.scatter(xs, ys, zs)
#add another set of data
#import pickle
with open('modi_tweets.pkl', 'rb') as f:
    modi_tweets = pickle.load(f)
from twitterexample import split_sentiments
modiX, modiY, modiZ = split_sentiments([get_sentiments(tweet_modi.text) for tweet_modi in modi_tweets])
ax.scatter(modiX, modiY, modiZ, color='r', marker='^')
ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('NARENDRA MODI')
plt.show()



from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')
sentiments = [get_sentiments(tweet_nag.text) for tweet_nag in nag_tweets]
# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])
ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))
ax.scatter(xs, ys, zs)
#add another set of data
import pickle
with open('nag_tweets.pkl', 'rb') as f:
    nag_tweets = pickle.load(f)
from twitterexample import split_sentiments
nagX, nagY, nagZ = split_sentiments([get_sentiments(tweet.text) for tweet in nag_tweets])
ax.scatter(nagX, nagY, nagZ, color='r', marker='^')
ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('AKKINENI NAGARJUNA')
plt.show()



from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')
sentiments = [get_sentiments(tweet_sachin.text) for tweet_sachin in sachin_tweets]
# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])
ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))
ax.scatter(xs, ys, zs)
#add another set of data
import pickle
with open('sachin_tweets.pkl', 'rb') as f:
    sachin_tweets = pickle.load(f)
from twitterexample import split_sentiments
sachinX, sachinY, sachinZ = split_sentiments([get_sentiments(tweet_sachin.text) for tweet_sachin in sachin_tweets])
ax.scatter(sachinX, sachinY, sachinZ, color='r', marker='^')
ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('SACHIN TENDULKAR')
plt.show()



plt.scatter(xs,zs,label='NARENDRA MODI')
plt.scatter(nagX,nagZ,color='r',marker='p',label='AKKINENI NAGARJUNA')
plt.scatter(sachinX,sachinZ,color='g',marker='d',label='SACHIN TENDULKAR')
plt.xlabel('negative')
plt.ylabel('positive')
plt.title('positive vs negative Sentiments')
plt.legend(loc='upper right',shadow=True)
plt.show()
```
